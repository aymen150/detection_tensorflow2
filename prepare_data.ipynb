```python
img = df["image"][4]
bboxes = df["rectangle"][4]
lmd.show_oneImage(img.copy(),bboxes)
bboxes
```

```python
### verticale augmentation
img_, bboxes_ = ca.RandomHorizontalFlip(1)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[0,0,0,0]])
bboxes_
```

```python
### scale augmentation
img_, bboxes_ = ca.RandomScale(0.1, diff = True)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
### translate augmentation
img_, bboxes_ = ca.RandomTranslate(0.3, diff = True)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
### ratation augmentation
img_, bboxes_ = ca.RandomRotate(25)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
### shear rotate agumentation
img_, bboxes_ = ca.RandomShear(0.1)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
### resize augmentation
img_, bboxes_ = ca.Resize(120)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
### color augmentation
img_, bboxes_ = ca.RandomHSV(100, 100, 100)(img.copy(), bboxes.copy())
lmd.show_oneImage(img_,bboxes_)
#lmd.show_oneImage(img_,[[1,1,1,1]])
bboxes_
```

```python
######################################################################################
#############################  DATA AUGMENTATION #####################################
######################################################################################

path_trainStopSignImg = pr.path_oneDataset("Traffic sign",tpe='test')[0]
path_trainStopSignLabel = path_trainStopSignImg + "Label/"
for i in range(len(df)): 
    """ ca.RandomShear(round(random.random(),2)),"""
    img_ = df["image"].iloc[i].copy()
    bboxes_ = df["rectangle"].iloc[i].copy()
    a = random.randint(1,3)
    try :
        for j in range(a):
                seq = ca.Sequence([ca.RandomHSV(random.randint(0,155),random.randint(0,155),random.randint(0,155)), ca.RandomHorizontalFlip(1), ca.RandomScale(round(random.uniform(0,0.20),2)), ca.RandomTranslate(random.uniform(0,0.3)), ca.RandomRotate(random.randint(0,10)),ca.Resize(random.randint(300,800))],random.uniform(0.2,1))
                img_, bboxes_ = seq(img_, bboxes_)
                nameFile = secrets.token_hex(16)
                print(nameFile)
                nameFileTxt = nameFile + ".txt"
                nameFileImg = nameFile + ".jpg"      
                if (ct.write_txt(label= "Stop sign",rectangle= bboxes_ ,nameFile = path_trainStopSignLabel+ nameFileTxt)):
                    cv2.imwrite(path_trainStopSignImg+nameFileImg,img_)
                
                #lmd.show_oneImage(img_,bboxes_)
    except :
                print("Error")
   # lmd.show_oneImage(img_,bboxes_)
 
```
